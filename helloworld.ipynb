```python
for i in ['日期', '时间', '生产线编号', '物料推送气缸推送状态', '物料推送气缸收回状态', '物料推送数', '物料待抓取数',
       '放置容器数', '容器上传检测数', '填装检测数', '填装定位器固定状态', '填装定位器放开状态', '物料抓取数', '填装旋转数',
       '填装下降数', '填装数', '加盖检测数', '加盖定位数', '推盖数', '加盖下降数', '加盖数', '拧盖检测数',
       '拧盖定位数', '拧盖下降数', '拧盖旋转数', '拧盖数', '合格数', '不合格数', '物料推送装置故障1001',
       '物料检测装置故障2001', '填装装置检测故障4001', '填装装置定位故障4002', '填装装置填装故障4003',
       '加盖装置定位故障5001', '加盖装置加盖故障5002', '拧盖装置定位故障6001', '拧盖装置拧盖故障6002']:
    print(i, )
    print(df[i].value_counts())
```

```python
#把故障变迁都转化为1
df['物料推送装置故障1001']= df['物料推送装置故障1001'].map(lambda x: 1 if x == 1001 else 0)
df['物料检测装置故障2001']= df['物料检测装置故障2001'].map(lambda x: 1 if x == 2001 else 0)
df['填装装置检测故障4001']= df['填装装置检测故障4001'].map(lambda x: 1 if x == 4001 else 0)
df['填装装置定位故障4002']= df['填装装置定位故障4002'].map(lambda x: 1 if x == 4002 else 0)
df['填装装置填装故障4003']= df['填装装置填装故障4003'].map(lambda x: 1 if x == 4003 else 0)
df['加盖装置定位故障5001']= df['加盖装置定位故障5001'].map(lambda x: 1 if x == 5001 else 0)
df['加盖装置加盖故障5002']= df['加盖装置加盖故障5002'].map(lambda x: 1 if x == 5002 else 0)
df['拧盖装置定位故障6001']= df['拧盖装置定位故障6001'].map(lambda x: 1 if x == 6001 else 0)
df['拧盖装置拧盖故障6002']= df['拧盖装置拧盖故障6002'].map(lambda x: 1 if x == 6002 else 0)
```

```python
#统计每种故障得出现次数
breakdown_cols=[col for col in df.columns if '故障' in col]
break_count=df[breakdown_cols].sum(axis=0)
print(break_count)
```

```python
# 设置支持中文的字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

plt.figure(figsize=(10,7))
plt.bar(break_count.index,break_count.values)
plt.title('故障统计')
plt.xlabel('故障类型')
plt.ylabel('故障次数')
plt.xticks(rotation=45) #将x轴标签旋转45度
```

```python
# 绘制故障趋势图
plt.figure(figsize=(40,10))
for col in breakdown_cols:
    plt.plot(df[col],label=col)
    plt.legend()
    plt.title('故障趋势图')
    plt.xlabel('时间')
    plt.ylabel('故障次数')
```
